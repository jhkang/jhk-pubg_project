```python
# 결측치 확인 함수
def check_missing_value(filename, df):
    # 결측치가 발생한 부분에서 log 파일(.csv) 생성
    if (df.isnull().sum()).sum() != 0:
        df.to_csv(f"./Example/Data/Error_log/{filename}.csv")
        raise Exception(f"Missing value: {(df.isnull().sum()).sum()}\nError_log: pubg_api/Example/Data/Error_log/{filename}.csv")
```

```python
# tournament_info 불러오기
tournament_info = pd.read_csv("./DB/tournament_info.csv", index_col = 0)
check_missing_value('tournament_info', tournament_info)

# match_info 불러오기
match_info = pd.read_csv("./DB/match_info.csv", index_col = 0)
check_missing_value('match_info', match_info)

# match_participant 불러오기
match_participant = pd.read_csv("./DB/match_participant.csv", index_col = 0)
check_missing_value('match_participant', match_participant)

data = match_participant

if os.path.isfile("./Example/Data/Train_data/train_match_data.csv"):
    # 해당 경로에 train_match_data.csv 파일이 있으면, 현재 내용을 해당 파일에 추가
    train_data = pd.read_csv(f"./Example/Data/Train_data/train_match_data.csv", index_col = 0)
    train_data = pd.concat([train_data, data], ignore_index=True)
    train_data.to_csv(f"./Example/Data/Train_data/train_match_data.csv")
else:
    # 해당 경로에 train_match_data.csv 파일이 없으면, 파일 생성
    data.to_csv(f"./Example/Data/Train_data/train_match_data.csv")
```

```python
data = pd.read_csv(f"./Example/Data/Train_data/train_match_data.csv", index_col = 0)
data.drop(["match_id", "player_id"], axis = 1, inplace = True)
```

```python
# VIF
z_label = (["dbnos", "assists", "boosts", "damage_dealt", "heals", "kill_streaks", "kills", "longest_kill", "revives", "ride_distance", "swim_distance", "headshot_kills", "vehicle_destroys", "walk_distance", "weapons_acquired"])
data_corr = data[z_label]

vif = pd.DataFrame()
vif["features"] = data_corr.columns
vif["VIF Factor"] = [variance_inflation_factor(data_corr.values, i) for i in range(data_corr.shape[1])]
vif
```

```python
# https://tjansry354.tistory.com/13

plt.rcParams['figure.figsize']=(15,10)
sb.heatmap(data_corr.corr(), annot=True, cmap='Blues', vmin = -1, vmax = 1)
```

```python
# linear regression -> round_point 획득량 예상 -> mvp 지표로 사용 가능
# https://tjansry354.tistory.com/12?category=927195

# 학습 데이터와 테스트 데이터로 분리합니다.
# 'round_point', 'win'를 제외한 칼럼들은 X
X = data[data.columns.difference(['round_point', 'win'])]
y = data['round_point']

# 검증(test)는 20%로 한다
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)

# 회귀 분석 객체 생성(선형 회귀 모델 생성)
lr = linear_model.LinearRegression()

# fit()는 기울기와 절편을 전달하기 위함.
lr_model = lr.fit(X_train, y_train)

x_new=X_test
y_new=lr_model.predict(x_new)

print("훈련 세트 정확도 : {:.3f}".format(lr_model.score(X_train,y_train)))
print("테스트 세트 정확도 : {:.3f}".format(lr_model.score(X_test,y_test)))

# 특성 중요도 시각화 하기
n_features = len(lr_model.feature_names_in_)
plt.barh(range(n_features), lr_model.coef_, align='center')
plt.yticks(np.arange(n_features), lr_model.feature_names_in_)
plt.xlabel("importances")
plt.ylabel("attr")
plt.ylim(-1, n_features)
plt.show()

# y_compare={'round_point_test':y_test, 'round_point_predicted':y_new}
y_compare={'y_test':y_test, 'y_predicted':y_new}
pd.DataFrame(y_compare)[:30].plot(y=['y_test', 'y_predicted'], alpha=0.6, kind="bar")
```

```python
# Random Forest -> 우승 예측
rf_data = standard_scaling(data)
z_label = (["dbnos", "assists", "boosts", "damage_dealt", "heals", "kill_streaks", "kills", "longest_kill", "revives", "ride_distance", "swim_distance", "headshot_kills", "vehicle_destroys", "walk_distance", "weapons_acquired"])
X = rf_data[z_label]
y = rf_data["win"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

forest = RandomForestClassifier(n_estimators=50, random_state=0)
forest.fit(X_train,y_train)

print("훈련 세트 정확도 : {:.3f}".format(forest.score(X_train,y_train)))
print("테스트 세트 정확도 : {:.3f}".format(forest.score(X_test,y_test)))

# 특성 중요도 시각화 하기
n_features = len(z_label)
plt.barh(range(n_features), forest.feature_importances_, align='center')
plt.yticks(np.arange(n_features), z_label)
plt.xlabel("importances")
plt.ylabel("attr")
plt.ylim(-1, n_features)
plt.show()
```

```python
mvp_feature = (forest.feature_importances_)*rf_data[z_label]
rf_data["win_feature"] = mvp_feature.sum(axis=1)

rf_data[['round_point','win','win_feature']].sort_values(["win_feature"], ascending=False)[:30]
```
